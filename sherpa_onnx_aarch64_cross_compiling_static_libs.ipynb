# Introduction

This colab notebook shows how to cross-compiling [sherpa-onnx](https://github.com/k2-fsa/sherpa-onnx) for aarch64.

It uses static libraries.

Please see
<https://k2-fsa.github.io/sherpa/onnx/install/aarch64-embedded-linux.html>
for more details.

# Install toolchain

````python
%%shell

mkdir -p $HOME/software
cd $HOME/software
wget -q https://huggingface.co/csukuangfj/sherpa-ncnn-toolchains/resolve/main/gcc-arm-10.3-2021.07-x86_64-aarch64-none-linux-gnu.tar.xz

tar xf gcc-arm-10.3-2021.07-x86_64-aarch64-none-linux-gnu.tar.xz
````

To check that we have installed the cross-compiling toolchain successfully, please run:



````python
%%shell

export PATH=$HOME/software/gcc-arm-10.3-2021.07-x86_64-aarch64-none-linux-gnu/bin:$PATH
aarch64-none-linux-gnu-gcc --version
````

Now we can build sherpa-onnx.

If you have the error

```
libtoolize: command not found
```

please run

````python
%%shell

sudo apt-get install -y libtool
````

````python
%%shell

export PATH=$HOME/software/gcc-arm-10.3-2021.07-x86_64-aarch64-none-linux-gnu/bin:$PATH

git clone https://github.com/k2-fsa/sherpa-onnx
cd sherpa-onnx
export BUILD_SHARED_LIBS=OFF
./build-aarch64-linux-gnu.sh
````

````python
%%shell

cd sherpa-onnx

ls -lh build-aarch64-linux-gnu/install/bin/
````

````python
%%shell

cd sherpa-onnx

readelf -d build-aarch64-linux-gnu/install/bin/sherpa-onnx
````

````python
%%shell

cd sherpa-onnx

tar cjvf build-aarch64-linux-gnu-static.tar.bz2 ./build-aarch64-linux-gnu/install

mv build-aarch64-linux-gnu-static.tar.bz2 ../
````

````python
%%shell

ls -lh build-aarch64-linux-gnu-static.tar.bz2
````
