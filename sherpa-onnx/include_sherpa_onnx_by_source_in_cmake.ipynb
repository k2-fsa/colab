# Introduction

This colab nodebook shows how to include sherpa-onnx by source in a CMake-based project.


```python
%%shell


mkdir $HOME/open-source
cd $HOME/open-source
mkdir sherpa-onnx-in-cmake-test
cd sherpa-onnx-in-cmake-test

git clone https://github.com/k2-fsa/sherpa-onnx

cat > main.cc <<EOF
#include "sherpa-onnx/c-api/cxx-api.h"

int main() {
  sherpa_onnx::cxx::OnlineTransducerModelConfig config;
  return 0;
}
EOF

cat > CMakeLists.txt <<EOF
cmake_minimum_required(VERSION 3.13 FATAL_ERROR)
project(sherpa-onnx-cmake)


list(APPEND CMAKE_MODULE_PATH \${CMAKE_SOURCE_DIR}/sherpa-onnx/cmake/Modules)
list(APPEND CMAKE_MODULE_PATH \${CMAKE_SOURCE_DIR}/sherpa-onnx/cmake)

include_directories(sherpa-onnx)

add_subdirectory(sherpa-onnx)

add_executable(main ./main.cc)
target_link_libraries(main sherpa-onnx-cxx-api)


EOF
```

```python
%%shell

cd  $HOME/open-source/sherpa-onnx-in-cmake-test
ls -lh

echo "---"

cat main.cc

echo "----"

cat CMakeLists.txt
```

```python
%%shell

cd  $HOME/open-source/sherpa-onnx-in-cmake-test

mkdir build
cd build
cmake ..
make -j2 main
```
