# Introduction

This colab notebook shows how to use models from
https://github.com/tjiho/French-tts-model-piper
with [sherpa-onnx](https://github.com/k2-fsa/sherpa-onnx)

```python
%%shell

pip install -q sherpa-onnx
git lfs install
git clone -q https://huggingface.co/csukuangfj/vits-piper-fr_FR-tjiho-model1
git clone -q https://huggingface.co/csukuangfj/vits-piper-fr_FR-tjiho-model2
git clone -q https://huggingface.co/csukuangfj/vits-piper-fr_FR-tjiho-model3
```

```python
%%shell

sherpa-onnx-offline-tts \
 --vits-model=./vits-piper-fr_FR-tjiho-model1/fr_FR-tjiho-model1.onnx \
 --vits-lexicon=./vits-piper-fr_FR-tjiho-model1/lexicon.txt \
 --vits-tokens=./vits-piper-fr_FR-tjiho-model1/tokens.txt \
 --output-filename=./test-model-1.wav \
 "après la pluie, le beau temps"

 sherpa-onnx-offline-tts \
 --vits-model=./vits-piper-fr_FR-tjiho-model2/fr_FR-tjiho-model2.onnx \
 --vits-lexicon=./vits-piper-fr_FR-tjiho-model2/lexicon.txt \
 --vits-tokens=./vits-piper-fr_FR-tjiho-model2/tokens.txt \
 --output-filename=./test-model-2.wav \
 "après la pluie, le beau temps"

 sherpa-onnx-offline-tts \
 --vits-model=./vits-piper-fr_FR-tjiho-model3/fr_FR-tjiho-model3.onnx \
 --vits-lexicon=./vits-piper-fr_FR-tjiho-model3/lexicon.txt \
 --vits-tokens=./vits-piper-fr_FR-tjiho-model3/tokens.txt \
 --output-filename=./test-model-3.wav \
 "après la pluie, le beau temps"
```
