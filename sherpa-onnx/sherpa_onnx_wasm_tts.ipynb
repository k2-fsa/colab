# WebAssembly for TTS (sherpa-onnx)

This colab notebook shows you how to run
https://k2-fsa.github.io/sherpa/onnx/tts/wasm/index.html

## Install Emscripten

```python
%%shell

git clone https://github.com/emscripten-core/emsdk.git
cd emsdk
git pull

# Please don't use the latest version below.
# Please see https://github.com/k2-fsa/sherpa-onnx/issues/1397
./emsdk install 3.1.64 # latest
./emsdk activate 3.1.64 # latest
source ./emsdk_env.sh

emcc -v
```

## Build

```python
%%shell

pushd ./emsdk
source ./emsdk_env.sh
popd


git clone https://github.com/k2-fsa/sherpa-onnx
cd sherpa-onnx

cd wasm/tts/assets

wget -q https://github.com/k2-fsa/sherpa-onnx/releases/download/tts-models/vits-piper-en_US-libritts_r-medium.tar.bz2
tar xf vits-piper-en_US-libritts_r-medium.tar.bz2
rm vits-piper-en_US-libritts_r-medium.tar.bz2
mv vits-piper-en_US-libritts_r-medium/en_US-libritts_r-medium.onnx ./model.onnx
mv vits-piper-en_US-libritts_r-medium/tokens.txt ./
mv vits-piper-en_US-libritts_r-medium/espeak-ng-data ./
rm -rf vits-piper-en_US-libritts_r-medium

cd ../../..

./build-wasm-simd-tts.sh
```

```python
%%shell

ls -lh
echo "---"
ls -lh sherpa-onnx/build-wasm-simd-tts/install/bin/wasm/tts
```

## Download the generated files to your local computer

```python
%%shell

mkdir -p sherpa-onnx-wasm-tts
cp -v sherpa-onnx/build-wasm-simd-tts/install/bin/wasm/tts/* ./sherpa-onnx-wasm-tts

tar cjvf ./sherpa-onnx-wasm-tts.tar.bz2 ./sherpa-onnx-wasm-tts
```
