# Introduction

This colab notebook shows how to cross-compiling [sherpa-onnx](https://github.com/k2-fsa/sherpa-onnx) for 32-bit ARM.

It uses static libraries.

Please see
<https://k2-fsa.github.io/sherpa/onnx/install/arm-embedded-linux.html>
for more details.

# Install toolchain

````python
%%shell

mkdir -p $HOME/software
cd $HOME/software
wget -q https://huggingface.co/csukuangfj/sherpa-ncnn-toolchains/resolve/main/gcc-arm-10.3-2021.07-x86_64-arm-none-linux-gnueabihf.tar.xz

tar xf gcc-arm-10.3-2021.07-x86_64-arm-none-linux-gnueabihf.tar.xz
````

To check that we have installed the cross-compiling toolchain successfully, please run:

````python
%%shell

export PATH=$HOME/software/gcc-arm-10.3-2021.07-x86_64-arm-none-linux-gnueabihf/bin:$PATH

arm-none-linux-gnueabihf-gcc --version
````

Now we can build sherpa-onnx.

If you have the error

```
libtoolize: command not found
```

please run

````python
%%shell

sudo apt-get install -y libtool
````

````python
%%shell

export PATH=$HOME/software/gcc-arm-10.3-2021.07-x86_64-arm-none-linux-gnueabihf/bin:$PATH

git clone https://github.com/k2-fsa/sherpa-onnx
cd sherpa-onnx
export BUILD_SHARED_LIBS=OFF
./build-arm-linux-gnueabihf.sh
````

````python
%%shell

cd sherpa-onnx

ls -lh build-arm-linux-gnueabihf/install/bin/
````

````python
%%shell

cd sherpa-onnx

readelf -d build-arm-linux-gnueabihf/install/bin/sherpa-onnx
````

````python
%%shell

cd sherpa-onnx

tar cjvf build-arm-linux-gnueabihf-static.tar.bz2 ./build-arm-linux-gnueabihf/install

mv build-arm-linux-gnueabihf-static.tar.bz2 ../
````

````python
%%shell

ls -lh build-arm-linux-gnueabihf-static.tar.bz2
````
