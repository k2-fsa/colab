# 4. Download the example source code

```python
%%shell
wget https://raw.githubusercontent.com/k2-fsa/sherpa-onnx/refs/heads/master/java-api-examples/NonStreamingTtsKittenEn.java
```

# 6. Download model files

```python
%%shell

curl -SL -O https://github.com/k2-fsa/sherpa-onnx/releases/download/tts-models/kitten-nano-en-v0_1-fp16.tar.bz2
tar xf kitten-nano-en-v0_1-fp16.tar.bz2
rm kitten-nano-en-v0_1-fp16.tar.bz2
```

```python
%%shell

# show model files
ls -lh kitten-nano-en-v0_1-fp16
```

# 7. Run it

```python
%%shell
ls -lh *.jar
ls -lh *.java
```

```python
%%shell

# For windows, please use ; to separate the two jar files
# For macOS and Linux, please use : to separate the two jar files
java -cp "./sherpa-onnx-v1.12.10.jar:./sherpa-onnx-native-lib-linux-x64-v1.12.10.jar"  NonStreamingTtsKittenEn.java
```
