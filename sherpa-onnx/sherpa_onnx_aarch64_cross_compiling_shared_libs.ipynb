# Introduction

This colab notebook shows how to cross-compiling [sherpa-onnx](https://github.com/k2-fsa/sherpa-onnx) for aarch64.

It uses shared libraries.

Please see
<https://k2-fsa.github.io/sherpa/onnx/install/aarch64-embedded-linux.html>
for more details.

# Install toolchain

````python
%%shell

mkdir -p $HOME/software
cd $HOME/software
wget https://huggingface.co/csukuangfj/sherpa-ncnn-toolchains/resolve/main/gcc-linaro-7.5.0-2019.12-x86_64_aarch64-linux-gnu.tar.xz
tar xf gcc-linaro-7.5.0-2019.12-x86_64_aarch64-linux-gnu.tar.xz
````

To check that we have installed the cross-compiling toolchain successfully, please run:



````python
%%shell

export PATH=$HOME/software/gcc-linaro-7.5.0-2019.12-x86_64_aarch64-linux-gnu/bin:$PATH

aarch64-linux-gnu-gcc --version
````

Now we can build sherpa-onnx.

If you have the error

```
libtoolize: command not found
```

please run

````python
%%shell

sudo apt-get install -y libtool
````

````python
%%shell

export PATH=$HOME/software/gcc-linaro-7.5.0-2019.12-x86_64_aarch64-linux-gnu/bin:$PATH


git clone https://github.com/k2-fsa/sherpa-onnx
cd sherpa-onnx
export BUILD_SHARED_LIBS=ON
./build-aarch64-linux-gnu.sh
````

````python
%%shell

cd sherpa-onnx

ls -lh build-aarch64-linux-gnu/install/bin/
````

````python
%%shell

cd sherpa-onnx

tar cjvf build-aarch64-linux-gnu.tar.bz2 ./build-aarch64-linux-gnu/install

mv build-aarch64-linux-gnu.tar.bz2 ../
````
