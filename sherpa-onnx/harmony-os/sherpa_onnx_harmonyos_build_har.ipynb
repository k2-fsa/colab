# Build sherpa-onnx for HarmonyOS

```python
%%shell

git clone https://github.com/k2-fsa/sherpa-onnx
```

## Build for arm64-v8a

```python
%%shell

cd sherpa-onnx
export OHOS_SDK_NATIVE_DIR=/content/command-line-tools/sdk/default/openharmony/native/

./build-ohos-arm64-v8a.sh
```

## Build for x86_64

```python
%%shell

cd sherpa-onnx
export OHOS_SDK_NATIVE_DIR=/content/command-line-tools/sdk/default/openharmony/native/

./build-ohos-x86-64.sh
```

## Build sherpa_onnx.har

```python
%%shell

cd sherpa-onnx/harmony-os/SherpaOnnxHar/

export PATH=/content/command-line-tools/bin:$PATH

hvigorw clean --no-daemon
hvigorw --mode module -p product=default -p module=sherpa_onnx@default assembleHar --analyze=normal --parallel --incremental --no-daemon

find . -name "*.har"
```

```python
%%shell

ls -lh /content/sherpa-onnx/harmony-os/SherpaOnnxHar/sherpa_onnx/build/default/outputs/default/sherpa_onnx.har
```
