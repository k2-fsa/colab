# Download pre-trained models

You can find a list of pre-trained models at
https://k2-fsa.github.io/sherpa/onnx/tts/pretrained_models/index.html

In the following, we show you two ways to download.

## Method 1

In this method, we use `git-lfs`:

```python
%%shell

# First, we need to install git-lfs
sudo apt-get install -y git-lfs
```

We will downlowd the following model
https://k2-fsa.github.io/sherpa/onnx/tts/pretrained_models/vits.html#ljspeech-english-single-speaker

```python
%%shell

GIT_LFS_SKIP_SMUDGE=1 git clone https://huggingface.co/csukuangfj/vits-ljs
cd vits-ljs
git lfs pull --include "*.onnx"
ls -lh
```

## Method 2

We will use
https://k2-fsa.github.io/sherpa/onnx/tts/pretrained_models/vits.html#vctk-english-multi-speaker-109-speakers
as an example.

```python
%%shell

# The model repo is at https://huggingface.co/csukuangfj/vits-vctk

mkdir vits-vctk
cd vits-vctk
wget https://huggingface.co/csukuangfj/vits-vctk/resolve/main/vits-vctk.onnx
wget https://huggingface.co/csukuangfj/vits-vctk/resolve/main/lexicon.txt
wget https://huggingface.co/csukuangfj/vits-vctk/resolve/main/tokens.txt
```

# Use of pre-trained models

Now we have downloaded two pre-trained models:

```python
%%shell

ls -lh
```

```python
%%shell

which sherpa-onnx-offline-tts
```

```python
%%shell

sherpa-onnx-offline-tts --help
```

We can also use the following file
https://github.com/k2-fsa/sherpa-onnx/blob/master/python-api-examples/offline-tts.py

```python
%%shell

wget https://raw.githubusercontent.com/k2-fsa/sherpa-onnx/master/python-api-examples/offline-tts.py
```

```python
%%shell

python3 ./offline-tts.py --help
```

```python
%%shell

python3 ./offline-tts.py \
 --vits-model=./vits-ljs/vits-ljs.onnx \
 --vits-lexicon=./vits-ljs/lexicon.txt \
 --vits-tokens=./vits-ljs/tokens.txt \
 --output-filename=./armstrong.wav \
 --debug=1 \
 "That's one small step for a man, a giant leap for mankind."
```

```python
from IPython.display import Audio, display
display(Audio("./armstrong.wav", autoplay=True))
```

```python
%%shell

# Change speech speed

python3 ./offline-tts.py \
 --vits-model=./vits-ljs/vits-ljs.onnx \
 --vits-lexicon=./vits-ljs/lexicon.txt \
 --vits-tokens=./vits-ljs/tokens.txt \
 --output-filename=./armstrong-2x.wav \
 --speed=2.0 \
 --num-threads=2 \
 "That's one small step for a man, a giant leap for mankind."
```

```python
from IPython.display import Audio, display
display(Audio("./armstrong-2x.wav", autoplay=True))
```

```python
%%shell

# Use the second model, but this time
# we use sherpa-onnx-offline-tts instead of offline-tts.py

sherpa-onnx-offline-tts \
  --vits-model=./vits-vctk/vits-vctk.onnx \
  --vits-lexicon=./vits-vctk/lexicon.txt \
  --vits-tokens=./vits-vctk/tokens.txt \
  --sid=0 \
  --output-filename=./kennedy-0.wav \
  --num-threads=2 \
  'Ask not what your country can do for you; ask what you can do for your country.'
```

```python
from IPython.display import Audio, display
display(Audio("./kennedy-0.wav", autoplay=True))
```

```python
%%shell

# We can change the speaker ID

sherpa-onnx-offline-tts \
  --vits-model=./vits-vctk/vits-vctk.onnx \
  --vits-lexicon=./vits-vctk/lexicon.txt \
  --vits-tokens=./vits-vctk/tokens.txt \
  --sid=99 \
  --output-filename=./kennedy-99.wav \
  --num-threads=2 \
  'Ask not what your country can do for you; ask what you can do for your country.'
```

```python
from IPython.display import Audio, display
display(Audio("./kennedy-99.wav", autoplay=True))
```

```python
%%shell

# We can still use offline-tts.py

python3 ./offline-tts.py \
  --vits-model=./vits-vctk/vits-vctk.onnx \
  --vits-lexicon=./vits-vctk/lexicon.txt \
  --vits-tokens=./vits-vctk/tokens.txt \
  --sid=88 \
  --output-filename=./kennedy-88.wav \
  --num-threads=2 \
  'Ask not what your country can do for you; ask what you can do for your country.'
```

```python
from IPython.display import Audio, display
display(Audio("./kennedy-88.wav", autoplay=True))
```
